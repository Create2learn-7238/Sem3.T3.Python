# try 
- **block let u test a block of code**

# except
- block let u handel  the error

# else
- let you execute the code if there is no error

# finally
- let u execute code , regardless of the result of try and except block

```python
def div(a,b):
    return a/b
try :
    print(div(3,0))
except :
    print("Error 404 : a/0 division by zero")
```

```python
def div(a,b):
    return a/b
try :
    print(div(1255,5))
except :
    print("Error : a/0 division by zero")
else :
    print("No Error")
finally :
    print("Always")
```
