# WAP to append ,delete and dispplay element of a list using class

```python
class list2:
    def __init__(self,l1):
        self.l1=l1
    def append(self,l2):
        self.l1.append(l2)
    def delete(self,r):
        self.l1.remove(r)
    def display(self):
        print(self.l1)
```

```python
l=list2([1,2,3])
while True:
    print("*******************************")
    print("Enter 1 to append")
    print("Enter 2 to delete")
    print("Enter 3 to display")
    print("Enter 0 to Exit ...")
    a=int(input("Enter your choice : "))
    if a==0:
        print("Exiting...")
        break
    elif a==1:
        a=int(input("Enter number to append : "))
        l.append(a)
    elif a==2:
        a=int(input("Enter number to delete : "))
        try :
            l.remove(a)
        except Exception as e:
            print("Error :", e)
    elif a==3:
        l.display()
    else :
        print("Invalid choice")
```

# WAP to create class atm and make methode to create pin change pin , check balance and widhraw and menu

```python
class atm:
    def __init__(self):
        self.bal=10000
    def createpin(self):
        pin = int(input("Enter 4 digit pin :"))
        if len(str(pin))==4:
            self.pin=pin
        else :
            print("Invalid enter 4 digit only")
        self.menu()
    def menu(self):
        print("********************************************")
        print("Choice :")
        print("1.Change pin")
        print("2.Widhraw")
        print("3.Check balance")
        print("0.Exit")
    def changepin(self):
        oldpin=int(input("Enter old Pin :"))
        if self.pin==oldpin :
            newpin=int(input("Enter New Pin :"))
            self.pin=newpin
        else : 
            print("Invalid Pin")
        self.menu()
    def checkbal(self):
        epin=int(input("Enter your Pin :"))
        if self.pin==epin :
            print(self.bal)
        self.menu()
    def widhraw(self):
        epin=int(input("Enter your Pin :"))
        if self.pin==epin :
            wid=int(input("Enter amt to widhraw :"))
            if wid<self.bal-2000 :
                self.bal=self.bal-wid
                print("Succeful widraw")
            else :
                print("No Balance")
        else :
            print("Invalid Pin")
        self.menu()
```

```python
a=atm()
c=1
a.createpin()
while c!=0:
    c=int(input("Enter your choice "))
    if c==1:
        a.changepin()
    elif c==2:
        a.widhraw()
    elif c==3:
        a.checkbal()
    elif c==0:
        print("Exi")
    else :
        print("Invalid")
```
