# WAP to count number of lines , words , character and spaces in a given files 

```python
#with loop
with open("Details.txt") as f:
    line=len(f.readlines())
    f.seek(0)
    word_count=0
    char=0
    for i in range(line):
        word=f.readline().split()
        for j in word :
            char=char+len(word)
            word_count=word_count+1  
    f.seek(0)
      
    print("Total line count : ",line)
    print("Total word count : ",word_count)
    print("Total char count",char)
```

```python
#without loop
with open("Details.txt") as f:
    b=f.read()
    x=b.split()
    print("Total word count : ",len(x))
    f.seek(0)
    print("Total line count : ",len(f.readlines()))
    y=len(b)-b.count(" ")-b.count("\n")
    print("Total char count",y)
    sp=b.count(" ")
    print("total spaces : ",sp)
```

# WAP to compare and display the difference

```python
with open("One.txt") as o:
    with open("Another.txt") as a:
        l=len(a.readlines())
        a.seek(0)
        for i in range(l):
            l1=a.readline()
            l2=o.readline()
            for j in range(len(l1)):
                if l1[j]!=l2[j] :
                    print("Line : ",i+1,"Pos : ",j)
```

```python
with open("code.txt") as c:
    with open("Code-without-coment.txt","w+") as cwc:
        l=len(c.readlines())
        c.seek(0)
        for i in range(l):
            line=c.readline()
            index=line.find("#")
            if index==-1:
                cwc.write(line)
            elif index==0:
                continue
            else:
                cwc.write(line[:index]+"\n")
        
print("Working...")       
```

# WAP to accept string from the user till the user enters END/0 each string enter by user should be a new line in text file save all the lines in text files and display only those lines which begins with capital letter

```python
with open("user.txt" , "w+") as f:
    print("Enter END to Stop")
    while True:
        line=input("Enter Lines to Write : ")
        if line=="END":
            break
        else:
            f.write(line+"\n")
    f.seek(0)
    a=len(f.readlines())
    f.seek(0)
    for i in range(a):
        x=f.readline()
        if x[0].isupper():
            print(x[:-1])
```

# WAP  to read line by line from a given line files f1 and f2 and write it into f3

```python
with open("f1.txt") as f1:
    with open("f2.txt")as f2:
        with open("f3.txt" , "w") as f3:
            x=len(f1.readlines())
            y=len(f2.readlines())
            d=0
            f1.seek(0)
            f2.seek(0)
            if x>y :
                d=y
            else :
                d=x
            for i in range(d):
                f3.write(f1.readline())
                f3.write(f2.readline())
            if d==x:
                f3.write(f2.read())
            if d==y:
                f3.write(f1.read())
        print("Working...")       
```

# WaP to read a text file and print The text file where the word starting with reverse 

```python
with open("reverse.txt") as f:
    data=f.read().split()
    new=""
    for i in data:
        if i[0].lower()=="i":
            new+=i[::-1]+" "
        else :
            new+=i+" "
            
    print(new)
```
